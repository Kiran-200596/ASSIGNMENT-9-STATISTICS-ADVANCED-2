**1).What is hypothesis testing in statistics?**

Hypothesis Testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

Hypothesis testing is of 2 types :
- Null Hypothesis : Means default assumption.
- Alternate Hypothesis : opposite of null hypothesis.

**2).What is the null hypothesis, and how does it differ from the alternative
hypothesis?**

**Null Hypothesis (Ho):**

- The default assumption in statistics.

- It assumes no effect, no difference, or no relationship exists.

- It’s what you try to disprove or test against.

- Example:
   - A company claims that the average weight of its chips packet is 50g
   - Null hypothesis(Ho) :mean is 50g

**Alternative Hypothesis (H1):**

- The statement we want to prove or support.

- It assumes there is an effect, difference, or relationship.

- It is accepted if Ho is rejected.

- Example:
   - A company claims that the average weight of its chips packet is 50g
   - Alternate hypothesis(H1) :mean is not 50g


**3). Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

The significance level (denoted as α) is the threshold probability we set before conducting a hypothesis test.
- It tells us how much risk we are willing to take of rejecting the null hypothesis (Ho) when it is actually true (Type I error).
- Common choices:
  - α = 0.05 (5%) → Most widely used
  - α = 0.01 (1%) → Stricter (used in medical trials, critical decisions)
  - α = 0.10 (10%) → More lenient

**Role in Decision Making**

- After running the test, we calculate a p-value.

- Decision Rule:
  - If p-value ≤ α → Reject Ho (Supports the alternate hypothesis)
  - If p-value > α → Fail to reject null hypothesis.

**Why is it Important?**

- Controls the probability of making a false positive (Type I error).

- Defines the strictness of the test.

- Helps balance risk between false positives (Type I error) and false negatives (Type II error).

**4).What are Type I and Type II errors? Give examples of each.**

- **Type I Error (False Positive):**

  - Definition: Rejecting the null hypothesis (Ho) when it is actually true.
  - Probability of making a Type I error = α (significance level).
  - Example :
    - Null Hypothesis: "A person does not have COVID."
    - Alternative: "The person has COVID."
    - If the test wrongly says positive when the person is actually healthy, that’s a Type I error.
    - Consequence: Unnecessary quarantine, treatment, or anxiety.

- **Type II Error (False Negative):**
  - Definition: Failing to reject the null hypothesis(Ho)when it is actually false.
  - Probability of making a Type II error = β.
  - The power of a test = 1−β.
  - Example:
    - Null Hypothesis: "A person does not have COVID."
    - Alternative: "The person has COVID."
    - If the test wrongly says negative when the person is actually infected, that’s a Type II error.
    - Consequence: Infected person goes untreated and spreads disease.

**5)What is the difference between a Z-test and a T-test? Explain when to use
each.**

- **Z-Test :**
  - Definition:
    - A statistical test used to determine whether there is a significant difference between sample and population means, or between two sample means, when population variance (σ²) is known.
  - Conditions to use Z-test:
    - Large sample size (n > 30).
    - Population standard deviation (σ) is known.
    - Data is approximately normally distributed.
  - Example:
    - Suppose the average weight of apples in a farm is known to be 150g with a population standard deviation of 10g.
    - If you take a large sample of 50 apples and want to test if the sample mean differs from 150g → **use Z-test.**

- **T-Test:**
  - Definition: A statistical test used when the sample size is small and/or population variance (σ²) is unknown. Instead of σ, we use the sample standard deviation (s).

  - Conditions to use T-test:

    - Small sample size (n ≤ 30).
    - Population standard deviation unknown.
    - Data is approximately normally distributed.

  - Example:
    - You collect exam scores of 20 students and want to test if the average score differs from 75.
    - Since the population variance is unknown and sample size is small → **use T-test.**

**6). Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for binomial distribution
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

**7).Implement hypothesis testing using Z-statistics for a sample dataset in
Python.Show the Python code and interpret the results.**

**sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

```python
import numpy as np
import scipy.stats as stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized mean
mu_0 = 50

# Sample statistics
x_bar = np.mean(sample_data)
sigma = np.std(sample_data, ddof=0)  # population-like SD
n = len(sample_data)

# Z statistic
z_stat = (x_bar - mu_0) / (sigma / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean = {x_bar:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.5f}")

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis: Mean is significantly different from 50")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from 50")
```

**8).Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Simulate data from normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=1000)  # mean=50, std=5, n=1000

# Step 2: Calculate mean and 95% confidence interval
mean = np.mean(data)
std_err = stats.sem(data)  # standard error = std/sqrt(n)
confidence = 0.95
ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=std_err)

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")

# Step 3: Plot histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci[0]:.2f}")
plt.axvline(ci[1], color='blue', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci[1]:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**9).Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Function to calculate Z-scores and plot histogram
    """
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = [(x - mean) / std for x in data]

    # Print summary
    print(f"Mean = {mean:.2f}, Standard Deviation = {std:.2f}")
    print("First 5 Z-scores:", z_scores[:5])

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=10, edgecolor='black')
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores


# Example dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

# Run function
z_scores = calculate_z_scores(data)
```
